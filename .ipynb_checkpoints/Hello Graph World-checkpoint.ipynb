### Objective

When software developers are starting to work in a new environment, they are encouraged to start with a simple “beachhead” program.  In this assignment, you’re asked to create the nodes and edges for a basic graph (for example, a Krackhardt kite).  

To complete this assignment Network X will be used.

### Create graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
#g = nx.krackhardt_kite_graph()
g = nx.generators.small.krackhardt_kite_graph()

print('Krackhardt Kite Nodes and Edges')
print(f'Nodes are {g.nodes()}')
print()
print(f'Edges are {g.edges()}')
```

```python
#Add labels
nodes_dict = {0:'Roger', 1:'Judie', 2:'Ken', 3:'Shannon', 4:'Lindsay', 5:'Cole', 6:'Cade', 7:'Digory', 8:'Louie', 9:'Theo'}

g = nx.relabel_nodes(g,nodes_dict)
```

```python
#Generate graph
plt.figure(figsize=(5,5))

g = nx.draw(g, with_labels=True)

plt.show()
```
